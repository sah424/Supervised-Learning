linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
X = [[1],[2],[3],[4],[5]] #hours studied
y = [40,50,65,75,90] #Marks
```

```python
model = LinearRegression() #Creating an object

model.fit(X,y) #passing the value

hours = float(input("How may hours you have studied : "))

predict_marks = model.predict([[hours]])

print(f"Based on your {hours}hour study in each day, you will obtain {predict_marks} marks")
```

logistic regression

Use LR when:

- Target is binary (0/1)
- Need probability output
- Data is linearly separable

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = [[1],[2],[3],[4],[5]] #hr studied
y = [0,0,1,1,1] #pass or fail -> 0 = Fail, 1 = Pass
```

```python
model = LogisticRegression() #Creating an object

model.fit(X,y) #Passing the value
# X = Independent Variable
# y = target Variable

hours = float(input("How many hours you have studied : "))

predict_result = model.predict([[hours]])[0]
# [[hours]] -> sklearn expect 2D Input
# [hours] -> It is 1D it will throw error
'''
[0] -> If we don't write this it will print array([1])/list[] -- but we want either 1 (pass) or 0(fail){} not [1] or [0]
Then we will compare it w 1 means coparing an array w a number which will throws an error
We use [0] to extract the actual predicted value.
'''


if predict_result == 1:
    print(f"Based on your {hours} study you will pass")
else :
    print(f"Based on your {hours} study you will fail")
```

KNN - K Nearest Neighbour

No learning of rules
â†’ Stores entire dataset
â†’ Looks at nearest K points
â†’ Majority vote of neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = [
    [180,7],
    [200, 7.5],
    [250, 8],
    [300,8.5],
    [330,9],
    [360,9.5]
]

# 0 = Apple, 1 = Orange
y = [0,0,0,1,1,1]

model = KNeighborsClassifier(n_neighbors = 3)

model.fit(X,y)

weight = float(input(f"Type the fruits weight in gram : "))
size = float(input(f"Type the fruits size in cm : "))

predictionss = model.predict([[weight, size]])[0]

if predictionss == 0:
    print("The fruit is Apple")
else :
    print ("The fruit is Orange")
```

Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X= [
    [7,2], #Apple
    [8,3], #Apple
    [9,8], #Orange
    [10,9] #Orange
]

y = [0,0,1,1] #0 = Apple , 1 = Orange

# Creating obj to pass the Algorighm
model = DecisionTreeClassifier()

model.fit(X,y)

# Creating Variable for input
size = float(input("Enter the fruit size in cm : "))
shade = float(input("Enter the color shade (1-10) : "))


# Creating another obj to predict using those variables
result = model.predict([[size,shade]])[0] #[0] used to get actual value other will we get a List 

if result == 0 : # 0 == Apple
    print("It is an Apple")
else : # 1 == Orange
    print("It is an Orange")
```

Random Forest

Learn decision boundaries from data
â†’ Each tree predicts based on splits
â†’ Final prediction = majority of trees

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Pass or Fail
X = [[10],[20],[36],[43],[50]] #Marks
y = [0,0,0,1,0] #Result // 0 = Fail, 1 = Pass

model = RandomForestClassifier()
model.fit(X,y) #Teaching model

Marks = float(input("Type your obtain marks : "))

result = model.predict([[Marks]])[0]

if result == 0:
    print("You Fail")
else :
    print("You Pass")
```
